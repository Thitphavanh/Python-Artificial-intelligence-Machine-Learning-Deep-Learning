```python
##.loc[:,[...]] ฟังก์ชันใช้เลือกเอาเฉพาะคอลัมน์ที่ต้องการนำมาใช้งาน
##.dropna() คำสั่งลบข้อมูล missing คือข้อมูลที่ไม่สมบูรณ์ออก
#data1.to_csv('D:/Face Detection/Datanew/data1.csv') #คำสั่งบันทึกไฟล์เก็บไว้เป็นนามสกุล .csv
dt = df.loc[:,['date','location','total_cases','new_cases','total_deaths','new_deaths']].dropna() #.set_index('date') #คำสั่ง .loc ใช้เลือกข้อมูลบางคอลัมน์
dt.tail(10) #เรียกข้อมูลมาจาก 10 แถวแรกท้ายสุด
```

```python
# # 10 Countries in AEC
# AEC = dt.loc[dt['location'].isin(['Singapore','Indonesia','Thailand','Myanmar','Vietnam','Brunei','Laos','Cambodia','Malaysia','Philippines'])] #.set_index('Entity') #.reset_index() 
# AEC

AEC = dt.loc[dt['location'].isin(['Laos'])].reset_index() #.set_index('date')
AEC
```

```python
plt.figure(figsize=(15,3))
sns.lineplot(AEC.index, 'new_cases', data=AEC, marker='o', markersize=10)
plt.title('COVID')
plt.show()
```

```python
X = AEC.index.values.reshape(-1,1) #ตัวแปรต้น
y = AEC.new_cases #.values.reshape(-1,1) #ตัวแปรตาม
print('แสดงข้อมูล 5 บรรทัดแรกของ X: \n',X[:5], '\nแสดงข้อมูล 5 บรรทัดแรกของ y: \n', y[:5]) #แสดงข้อมูล 5 บรรทัดแรกของ X และ y
```

```python
y
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=3)
x_poly = poly_features.fit_transform(X)
model = LinearRegression()
model.fit(x_poly, y)
```

```python
y_poly_pred = model.predict(x_poly)
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = 15,3
plt.scatter(X,y,color='b',s=12)
plt.plot(X,y_poly_pred, linewidth='2',color='r')
plt.ylabel('New Cases')
plt.title('Degree=3')
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np
print('R2 = {:.5f}'.format(r2_score(y, y_poly_pred)))
```

```python
#ค่าเฉลี่ยความคลาดเคลื่อนยกกำลังสอง
# ถ้าค่า MSE เข้าใกล้ 0 จะดี แต่ถ้าาเข้าใกล้ 1 ไม่ค่อยดี 
print('ค่าเฉลี่ยความคลาดเคลื่อนยกกำลังสอง', mean_squared_error(y,y_poly_pred).round(2))
```

```python
#ค่าเฉลี่ยความคลาดเคลื่อนสัมบูรณ์
# ถ้าค่า MAE เข้าใกล้ 0 จะดี แต่ถ้าาเข้าใกล้ 1 ไม่ค่อยดี 
print('ค่าเฉลี่ยความคลาดเคลื่อนสัมบูรณ์', mean_absolute_error(y,y_poly_pred).round(2))
```
