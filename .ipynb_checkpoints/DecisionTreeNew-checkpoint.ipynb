```python
import pandas as pd
data = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
data
```

```python
df = data.loc[:,['total_cases','location']].dropna()
df
```

```python
s = df.loc[df['location'].isin(['Thailand','Laos','Cambodia'])].reset_index()
st = s[['total_cases','location']]
st
```

```python
X = st[['total_cases']].values
y_names = st[['location']].values
print('X_Features',X,'\n\ny_Features',y_names)
```
