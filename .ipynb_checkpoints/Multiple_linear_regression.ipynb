```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# formula: response ~ predictor1 + predictor2
model_a = smf.ols(formula='total_deaths ~ total_cases + new_cases + gdp_per_capita + life_expectancy', data=train).fit()
# model_a = smf.ols(formula='Sales ~ TV + Radio', data=df).fit()
```

```python
print(model_a.summary())
```

### ถ้าค่า p (P>|t|) < 0.05 แสดงว่ามีความสัมพันธ์กัน คือคือตัวแปรต้น (X หรือเหตุ) เป็นสาเหตุ/ส่งผลต่อตัวแปรตาม (Y หรือผล) เช่น กินจุ (เหตุ) ทำให้อ้วน (ผล) ขณะเดียวกัน อ้วน (เหตุ) ทำให้เป็นโรคความดัน เบาหวาน (ผล) เป็นต้น
### ถ้าค่า p (P>|t|) > 0.05 แสดงว่าไม่มีความสัมพันธ์กัน คือคือตัวแปรต้น (X หรือเหตุ) ไม่เป็นสาเหตุ/ไม่ส่งผลต่อตัวแปรตาม (Y หรือผล) เช่น กินจุ (เหตุ) แต่ก็ไม่ทำให้อ้วน (ผล) ขณะเดียวกัน อ้วน (เหตุ) ไม่ทำให้เป็นโรคความดัน เบาหวาน (ผล) เป็นต้น
### ค่าอย่างมีนัยสำคัญทางสถิติที่..... ถ้าค่า p < 0.05 แต่ > 0.01 อย่างมีนัยสำคัญทางสถิติที่ 0.05, ถ้าค่า p < 0.01 แต่ > 0.001 อย่างมีนัยสำคัญทางสถิติที่ 0.01, ถ้าค่า p < 0.001 อย่างมีนัยสำคัญทางสถิติที่ 0.000

### การอธิบายค่าสัมประสิทธิ์ถดถอย (coef) ถ้าค่าเป็นบวก (18.7179) หมายความว่า x1 (ตัวแปรอิสระ/ต้น: เหตุ) เพิ่มขึ้น ส่งผลให้ Y (ตัวแปรตาม: ผล) เพิ่มตามไปด้วย ขณะเดียวกัน ถ้าค่าเป็นลบ (-22.6050) หมายความว่า x1 (ตัวแปรอิสระ: เหตุ) เพิ่มขึ้น ส่งผลให้ Y (ตัวแปรตาม: ผล) ลดลง แต่ค่าลบในที่นี้ไม่ต้องนำมาอธิบาย เพราะไม่มีนัยสำคัญ คือค่า p > 0.05

### แบบจำลองในการศึกษา 
### Y = a + b1x1 + b2x2 + b3x3 + e
### กำหนดให้
#### a = ค่าคงที่ความถดถอยพหุ
#### b1 ... b5 = ค่าสัมประสิทธิถดถอยของตัวแปรอิสระ (ตัวแปรต้น)
#### Y = total_deaths
#### x1 = aged_65_older
#### x2 = aged_70_older
#### x3 = life_expectancy
#### e = ค่าความคลาดเคลื่อน

### การอธิบายแปลผล เอาเฉพาะตัวที่มีนัยสำคัญทางสถิติ คือค่า p < 0.05 เท่านั้นมาอธิบายแปลผล
### ผลการศึกษาพบว่า aged_65_older (x1) มีความสัมพันธ์ในทิศทางบวกกับ total_deaths (Y) ที่ระดับนัยสำคัญทางสถิติ 0.000 โดยมีค่าสัมประสิทธิ์ aged_65_older เท่ากับ 1.174e+04 หมายความว่า ถ้า aged_65_older เพิ่มขึ้น 1 หน่วย ส่งผลทำให้ total_deaths เพิ่มขึ้น 1.174e+04 หน่วย 
### aged_70_older (x2) มีความสัมพันธ์ในทิศทางลบกับ total_deaths (Y) ที่ระดับนัยสำคัญทางสถิติ 0.000 โดยมีค่าสัมประสิทธิ์ aged_70_older เท่ากับ -1.602e+04 หมายความว่า ถ้า aged_70_older เพิ่มขึ้น 1 หน่วย ส่งผลทำให้การ total_deaths ลดลง 1.602e+04 หน่วย
