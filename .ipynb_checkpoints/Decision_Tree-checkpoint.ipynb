```python
## pd.read_csv การดึงข้อมูลตารางออกมาแล้วนำไปเก็นไว้ในตัวแปร df
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data/Supervised_Learning_Classification/owid-covid-data.csv',parse_dates=['date']) #, index_col='Entity', parse_dates=True)
df
```

```python
## .info() คำสั่ง method แสดงประเภทรายการข้อมูลว่าเป็นตัวเลขหรือ string
df.info()
```

```python
## .shape คำสั่งใช้แสดงจำนวนคอลัมน์และแถว (นำนวนแถว, จำนวนคอลัมน์)
df.shape
```

```python
#.drop([...]) ฟังก์ชันลบคอลัมน์ที่ไม่ต้องการออกไป
#df.to_csv('D:/Python books/Datanew/data1.csv')
#CD = covid.drop([21,22,23,24]) #,axis=0) #คำสั่งใช้ลบแถวที่ไม่ต้องการออกไป
data = df.drop(['weekly_icu_admissions','weekly_icu_admissions_per_million','weekly_hosp_admissions','weekly_hosp_admissions_per_million'], axis=1) #ลบคอลัมน์
data
```

```python
##.loc[:,[...]] ฟังก์ชันใช้เลือกเอาเฉพาะคอลัมน์ที่ต้องการนำมาใช้งาน
##.dropna() คำสั่งลบข้อมูล missing คือข้อมูลที่ไม่สมบูรณ์ออก
#data1.to_csv('D:/Face Detection/Datanew/data1.csv') #คำสั่งบันทึกไฟล์เก็บไว้เป็นนามสกุล .csv
dt = data.loc[:,['total_cases','new_cases','total_deaths','new_deaths','location']].dropna() #คำสั่ง .loc ใช้เลือกข้อมูลบางคอลัมน์
dt
```

```python
#dt['location'].unique() คำสั่งแสดงรายชื่อประเทศว่ามีประเทศอะไรบ้าง 
b = dt.location.unique() # .unique() เป็น method ใช้ดูรายชื่อประเทศว่ามีประเทศอะไรบ้าง
b
```

```python
##len() คำสั่งนับจำนวนประเทศว่ามีกี่ประเทศ
len(b)
```

```python
# เลือกข้อมูลมา 5 ประเทศ
#sn.to_csv('/content/drive/MyDrive/Colab Notebooks/Data/Supervised_Learning_Classification/covid-dataNew.csv')
sn = dt.loc[dt['location'].isin(['Singapore','Indonesia','Thailand','Myanmar','Malaysia'])] #.set_index('location') #.reset_index()
sn
```

```python
#dt['location'].unique() คำสั่งแสดงรายชื่อประเทศว่ามีประเทศอะไรบ้าง 
c = sn.location.unique() # .unique() เป็น method ใช้ดูรายชื่อประเทศว่ามีประเทศอะไรบ้าง
c
```

```python
## LabelEncoder() เป็นคำสั่งแปลงชื่อประเทศให้เป็นเลขอันดับ (sequences) เช่น 0 1 2 3 4 เป็นต้น
## Import label encoder 
#from sklearn import preprocessing
## label_encoder object knows how to understand word labels. 
#label_encoder = preprocessing.LabelEncoder()
## Encode labels in column 'location'. 
#sn['location']= label_encoder.fit_transform(sn['location']) 
#print(sn.head())
```

```python
## การแบ่งข้อมูลเพื่อ train โมเดล ให้อยู่ในตัวแปรต้น (X) และตัวแปรตาม (y)
#X = df1[:,0:4]
#y = df1[:,4]
#print('X',y)
#print('y',y)
```

```python
## การแบ่งข้อมูลอีกวิธีหนึ่ง เพื่อ train โมเดล ให้อยู่ในตัวแปรต้น (X) และตัวแปรตาม (y)
## .values การแบ่งข้อมูลแล้วแปลงให้เป็นข้อมูลอาเรย์
X = sn.drop('location',axis=1).values 
y = sn['location'].values 
print('X:', X)
print('y:', y)
```

```python
## StandardScaler คำสั่งแปลงข้อมูลในตัวแปร X ให้เป็น binary คือ 0 1
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X)
X = sc.transform(X)
print('X:',X)
```

```python
# การแบ่งข้อมูลเพื่อนำไปสู่การ train และ test โมเดล ในที่นี่ใช้ train set 70% และใช้ test set 30%
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=2020)
```

```python
clf = tree.DecisionTreeClassifier(criterion = 'gini', random_state=0)
clf.fit(X_train, y_train)
```

```python
# generate evaluation metrics
import sklearn.metrics as metrics
from sklearn.metrics import classification_report,confusion_matrix

print("Train - Accuracy :", metrics.accuracy_score(y_train, clf.predict(X_train)))
print("Train - Confusion matrix :\n",metrics.confusion_matrix(y_train, clf.predict(X_train)))
print("Train - classification report :\n", metrics.classification_report(y_train, clf.predict(X_train)))
```

```python
import sklearn.metrics as metrics
from sklearn.metrics import classification_report,confusion_matrix

print("Test - Accuracy :", metrics.accuracy_score(y_test, clf.predict(X_test)))
print("Test - Confusion matrix :\n",metrics.confusion_matrix(y_test, clf.predict(X_test)))
print("Test - classification report :\n", metrics.classification_report(y_test, clf.predict(X_test)))
```

```python
## การสร้างต้นไม้ตัดสินใจ
# Create Decision Tree classifer object 
clf = DecisionTreeClassifier(criterion="gini", max_depth=None) #"entropy"

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

## การทำนายต้นไม้ตัดสินใจ
#Predict the response for test dataset
y_pred = clf.predict(X_test)
y_score = clf.score(X,y)
```
