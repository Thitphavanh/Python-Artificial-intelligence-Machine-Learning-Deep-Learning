```python
class Dog:
    # (Attributes/Features)
    eyes = 2
    ears = 2
    nose = 2
    # (Constructor/Initializer)

    def __init__(self, eyes, ears):
        self.eyes = eyes
        self.ears = ears

    # (Behavior/Method)
    def eye(self):
        e1 = self.eyes * 10
        return e1

    def ear(self):
        e2 = self.ears * 20
        return e2


data = Dog(50, 2)
print(f'ໄລຍະການເຫັນໃນຕາຂອງໝາ ແມ່ນ {data.eye()} Km')
print(f'ໄລຍະການໄດ້ຍິນໃນຫູຂອງໝາ ແມ່ນ {data.ear()} Km')
```
